```python
import pandas as pd
```

```python
#Question a
#Read the csv file named it dataset
dataset = pd.read_csv('~/Downloads/890/student_results.csv')
```

```python
dataset
```

```python
#Question b
# replace NA value with zeroes
dataset = dataset.fillna(0)
dataset
```

```python
#Question c
#calculate the mark by dividing the total and multiply by the Weight
Assessment = dataset["Assessment 1"]/70*0.07 + dataset["Assessment 2"]/30*0.03 + dataset["Assessment 3"]/70*0.07 + dataset["Assessment 4"]/30*0.03
Online_Quiz = dataset["Online quiz 1 "]/30*0.03 + dataset["Online quiz 2"]/20*0.02 + dataset["Online quiz 3"]/30*0.03 + dataset["Online quiz 4"]/20*0.02
Group_Assignment = dataset["Group assignment 1"]/30*0.024 + dataset["Group assignment 2"]/20*0.016 + dataset["Group assignment 3"]/50*0.04
Group_evaluation = dataset[" Group evaluation"]/2*0.02
Final_exam = dataset["Final exam"]/100*0.6
#add up all the weight marks and multiply by 100, and create a new column call Total.marks
dataset["Total.marks"] = (Assessment + Online_Quiz + Group_Assignment + Group_evaluation + Final_exam)*100
dataset
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#use the module seaborn to plot the boxplot
Q3bBoxplot = sns.boxplot(x="Student category", y = "Total.marks", data = dataset)
```

```python
#Question d
#Calculate the mean, var and std of total marks for each class
mean = dataset[["Class No", "Total.marks"]].groupby('Class No').mean()
Variance = dataset[["Class No", "Total.marks"]].groupby('Class No').var()
Standard_deviation = dataset[["Class No", "Total.marks"]].groupby('Class No').std()
# Merge 3 table in to one stat_table by pandas
stat_table = pd.concat([mean, Variance, Standard_deviation], axis = 1)
#Change the colunms name of the table
stat_table.columns = ['Mean','Variance', 'Standard_deviation']
stat_table
                        
```

```python
#Question e
#Give Grade to each student
Grade = list()

for marks in dataset['Total.marks']:
#Setting difference condition by if function
    if 0<= marks < 50:
        Grade.append('F')
    elif 50<= marks < 65:
        Grade.append('P')
    elif 65<= marks < 75:
        Grade.append('Cr')
    elif 75<= marks < 85:
        Grade.append('D')
    else:
        Grade.append('HD')
#create new colunm and equal to Grade,        
dataset['Grade'] = Grade
#print out the result
dataset
```

```python
#Question f
import statsmodels.api as sm
#predictor
X = dataset["Assessment 1"]
#response
y = dataset["Final exam"]
#add a constant term in the regression
X = sm.add_constant(X)
#fit a simple regression model
model = sm.OLS(y, X).fit()
#make predictions by the model
predictions = model.predict(X)
#print out statistic result
model.summary()
```
